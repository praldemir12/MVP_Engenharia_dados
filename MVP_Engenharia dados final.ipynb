### 7.3. Reunião de Dados Anuais

### Descrição
O objetivo deste código é reunir dados de tabelas que contêm registros dos anos de 2018 a 2022 e combiná-los em um único DataFrame Pandas, utilizando Apache Spark. Estas tabelas possuem o mesmo esquema, mas contêm dados de diferentes anos. Ao centralizar esses dados, facilitamos análises e visualizações subsequentes.

### Procedimento
1. **Inicialização da SparkSession**: Importamos a SparkSession do módulo pyspark.sql. Esta é a principal porta de entrada para a funcionalidade do Spark SQL.
2. **Criação ou Recuperação da Sessão**: Usamos `SparkSession.builder.getOrCreate()` para iniciar ou recuperar uma sessão Spark existente.
3. **Definição dos Anos de Interesse**: Estabelecemos os anos de início e fim (2018-2022) que desejamos consolidar.
4. **Preparação da Lista de DataFrames**: Uma lista vazia, `lista_dfs`, é inicializada para armazenar os DataFrames correspondentes a cada ano.
5. **Looping por Anos**: Iteramos por cada ano no intervalo especificado.
    - Dentro deste loop, construímos dinamicamente o nome da tabela para o ano atual.
    - Uma consulta SQL é executada para selecionar todos os registros da tabela do ano em questão.
    - O DataFrame resultante é então adicionado à nossa lista.
6. **Unificação de DataFrames**: Uma vez que todos os anos foram processados, usamos a operação `union` para combinar os DataFrames individuais em um único DataFrame, `df_unificado`.
7. **Conversão para DataFrame Pandas**: O DataFrame Spark unificado é convertido em um DataFrame Pandas, `df_combinado`.
8. **Visualização**: As primeiras linhas do DataFrame Pandas resultante são exibidas para verificação.

A seguir, segue-se a implementação detalhada do código.


```
# Criar ou obter a sessão Spark existente
spark = SparkSession.builder.getOrCreate()

# Defina o intervalo de anos que deseja ler
inicio_ano = 2018
fim_ano = 2022  # Ajuste conforme necessário

# Lista para armazenar os DataFrames
lista_dfs = []

# Ler as tabelas para cada ano e adicionar ao lista_dfs
for ano in range(inicio_ano, fim_ano + 1):
    tabela_nome = f"`dt0028_dev`.`default`.`escolas_atendidas_{ano}`"
    df = spark.sql(f"SELECT * FROM {tabela_nome}")
    lista_dfs.append(df)

# Se desejar, você pode combinar todos os DataFrames em um único DataFrame
df_unificado = lista_dfs[0]
for df in lista_dfs[1:]:
    df_unificado = df_unificado.union(df)

#Converter para o Pandas e mostrar as primeiras linhas do dataframe
df_combinado = df_unificado.toPandas()
df_combinado
```

### 8.1. Distribuição da Quantidade de Escolas Atendidas com Merenda Escolar por Estado

### Objetivo
Compreender a distribuição anual da quantidade de escolas que foram beneficiadas com merenda escolar em cada estado.

### Metodologia

- **Importação das Bibliotecas**: As bibliotecas `matplotlib.pyplot` e `seaborn` são essenciais para a visualização dos dados.

- **Configuração do Estilo do Gráfico**: Usando `sns.set(style="whitegrid")`, estabelecemos um estilo de fundo que facilita a visualização dos dados.

- **Filtragem dos Dados**: Extraímos as colunas 'Ano', 'UF' e 'QtdEscolasAtendidas' do DataFrame principal para focar nas informações necessárias.

- **Agrupamento e Soma**: Utilizamos a função `groupby()` para agrupar os dados filtrados por 'Ano' e 'UF'. Em seguida, calculamos a soma da 'QtdEscolasAtendidas' para cada combinação, criando o DataFrame `dados_agrupados`.

- **Ordenação dos Estados**: Com base na quantidade total de escolas atendidas, os estados foram ordenados. Esta ordenação foi feita usando a função `sort_values()` e será utilizada para a plotagem do gráfico.

### Visualização

Após a preparação, um gráfico foi gerado para representar os dados. A visualização permite responder à pergunta inicial e compreender melhor a distribuição da merenda escolar entre os estados ao longo dos anos considerados.


```
# Definindo o estilo do gráfico
sns.set(style="whitegrid")

# Filtrando os dados para incluir apenas as colunas relevantes ('Ano', 'UF' e 'QtdEscolasAtendidas')
dados_filtrados = df_combinado[['Ano', 'UF', 'QtdEscolasAtendidas']]

# Agrupando os dados por 'Ano' e 'UF', e somando a 'QtdEscolasAtendidas'
dados_agrupados = dados_filtrados.groupby(['Ano', 'UF']).sum().reset_index()

# Definindo a ordenação dos estados por quantidade total de escolas atendidas
order = dados_agrupados.groupby('UF')['QtdEscolasAtendidas'].sum().sort_values(ascending=False).index

# Plotando o gráfico de barras
# Plotando o gráfico de barras
plt.figure(figsize=(15, 8))
sns.barplot(data=dados_agrupados, x='UF', y='QtdEscolasAtendidas', hue='Ano', order=order)

# Exibindo o gráfico
plt.show()
```

O principal objetivo deste gráfico é abordar a seguinte questão:

> **Quais estados apresentaram uma diminuição no número de escolas atendidas com merenda em qualquer ano durante o período entre 2018 e 2022?**

### Processamento e Análise dos Dados

Para esta análise, utilizamos a base de dados `df_combinado`. O processo de preparação e análise dos dados consiste nas seguintes etapas:

1. **Preparação dos Dados**: Os dados são primeiramente agrupados por 'Ano' e 'Estado'. A seguir, é calculada a diferença no número de escolas atendidas entre 2022 e 2018 para cada estado.
2. **Seleção dos Top 10**: Após o cálculo das diferenças, os estados são ordenados com base nesta métrica. Os 10 estados que apresentaram a maior diminuição no número de escolas atendidas são então selecionados para visualização.
3. **Visualização**: Para representar os dados, um gráfico de linhas é gerado. A análise resulta em um gráfico elucidativo que destaca os 10 estados que sofreram a maior diminuição no número de escolas atendidas com merenda escolar entre 2018 e 2022.


```
# Supondo que você já tenha df_combinado definido em algum lugar do seu código:
dados_estados = df_combinado.groupby(['Ano', 'UF'])['QtdEscolasAtendidas'].sum().unstack('Ano').reset_index()

# Calculando a diferença entre 2022 e 2018 para cada estado
dados_estados['Diferença'] = dados_estados[2022] - dados_estados[2018]

# Ordenando os estados pela maior diminuição e selecionando os top 10
top_diminuicao = dados_estados.sort_values(by='Diferença').head(10)

# Plotando o gráfico
plt.figure(figsize=(15, 10))

for index, row in top_diminuicao.iterrows():
    anos = row.index[1:-1].astype(str)  
    valores = row.values[1:-1]  
    
    plt.plot(anos, valores, marker='o', label=row['UF'])
    
    # Colocando a sigla do estado na frente da linha (com um deslocamento maior)
    plt.annotate(row['UF'], 
                 (anos[0], valores[0]), 
                 textcoords="offset points", 
                 xytext=(-25,0),
                 ha='center',
                 fontsize=10,
                 color='black',
                 weight='bold')
    
    # Adicionando números nos pontos
    for ano, valor in zip(anos, valores):
        plt.annotate(str(int(valor)), 
                     (ano, valor), 
                     textcoords="offset points", 
                     xytext=(0,5), 
                     ha='center', 
                     fontsize=9)

plt.title('Top 10 Estados com Maior Diminuição no Número de Escolas Atendidas com Merenda (2018-2022)')
plt.xlabel('Ano')
plt.ylabel('Número de Escolas Atendidas')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

### 12.2. Processamento e Limpeza de Dados em Conjuntos de Dados Municipais

O código apresentado realiza diversas operações de processamento e limpeza de dados em dois conjuntos de dados, com o objetivo final de unir esses conjuntos.

## Carregamento dos Dados

O código inicia carregando um conjunto de dados nomeado `municipios_2021`, que contém informações sobre municípios brasileiros no ano de 2021.

## Operações de Normalização

Em seguida, o código prepara uma série de operações de normalização para ambos os conjuntos de dados - `municipios_2021` e outro chamado `df_2021` (informações sobre escolas atendidas por município em 2021). Essas operações de normalização incluem:

1. **Remoção de acentos:** Todos os acentos nos nomes dos municípios são removidos, utilizando uma lista de caracteres acentuados e seus equivalentes sem acento.
2. **Conversão para letras minúsculas:** Todos os caracteres nos nomes dos municípios são convertidos para letras minúsculas.
3. **Remoção de caracteres especiais e espaços:** Qualquer caractere que não seja uma letra minúscula é removido dos nomes dos municípios.

## União dos Conjuntos de Dados

Após as operações de normalização em ambos os conjuntos de dados, o código realiza uma operação de "join" (união) entre eles. A união é feita com base nos nomes dos municípios, após terem sido normalizados, de ambos os conjuntos de dados.

## Resumo

Em resumo, o código prepara e une dois conjuntos de dados baseados em nomes de municípios, após realizar um processo de normalização para garantir que as correspondências sejam precisas e que variações devido a acentos, capitalização ou caracteres especiais não impeçam a união correta dos registros.


```
# Carregar o dataframe da tabela 'a'
df_2021 = spark.sql("SELECT * FROM `dt0028_dev`.`default`.`escolas_atendidas_2021`;")

# Carregar o dataframe da tabela 'a'
df_spark = spark.sql("SELECT * FROM `dt0028_dev`.`default`.`municipios_2021`")

# Lista de acentos
accents = "áÁâÂàÀäÄãÃéÉêÊëËíÍîÎïÏóÓôÔöÖõÕúÚûÛüÜçÇ"
no_accents = "aAaAaAaAaAeEeEeEiIiIiIoOoOoOoOuUuUuUcC"

# Normalize o dataframe df_spark (tabela 'a')
df_spark_normalized = (df_spark.withColumn("normalized_name", 
                 lower(translate(col("name"), accents, no_accents))
                 .alias("normalized_name")))

# Removendo caracteres especiais e espaços do df_spark_normalized
df_spark_normalized = df_spark_normalized.withColumn("normalized_name", regexp_replace(col("normalized_name"), "[^a-z]", ""))

# Normalize o dataframe df_2021 (tabela 'b')
df_2021_normalized = (df_2021.withColumn("normalized_municipio", 
                 lower(translate(col("Municipio"), accents, no_accents))
                 .alias("normalized_municipio")))

# Removendo caracteres especiais e espaços do df_2021_normalized
df_2021_normalized = df_2021_normalized.withColumn("normalized_municipio", regexp_replace(col("normalized_municipio"), "[^a-z]", ""))

# Join entre os dataframes
joined_data = df_spark_normalized.join(df_2021_normalized, df_spark_normalized.normalized_name == df_2021_normalized.normalized_municipio)
```

### 12.8. Relatório de Análise da Relação entre Escolas Atendidas com Merenda Escolar e População dos Estados Brasileiros

### Objetivo

O relatório tem o objetivo de analisar e visualizar a relação percentual entre a quantidade de escolas atendidas com merenda escolar e a população dos estados brasileiros.

### Metodologia

1. **Agrupamento por Estado:** Os dados são agrupados pelo código de estado (`uf_code`).
2. **Agregação:** Uma agregação é realizada para somar a população e a quantidade de escolas atendidas de cada estado.
3. **Cálculo da Relação Percentual:** A relação percentual entre a quantidade de escolas atendidas e a população total de cada estado é calculada.
4. **Ordenação:** Os estados são ordenados com base nessa relação percentual, do maior para o menor.
5. **Conversão para DataFrame:** Os dados processados são convertidos para um DataFrame do pandas.
6. **Visualização:** Um gráfico de barras é gerado usando a biblioteca `matplotlib`. O eixo y do gráfico é formatado para representar os valores em porcentagem.

### Resultados

O gráfico resultante permite visualizar, para cada estado (UF), qual é o percentual de escolas atendidas com merenda escolar em relação à sua população.

### Insights

A visualização fornece insights sobre como a densidade populacional de um estado pode se relacionar com o atendimento de merenda nas escolas.



```
# Agrupando por uf_code e somando a população e a quantidade de escolas atendidas
state_data = joined_data.groupBy('uf_code').agg({'pop_21':'sum', 'QtdEscolasAtendidas':'sum'})

# Calculando a relação percentual entre a quantidade de escolas atendidas e a população por estado
state_data = state_data.withColumn('percent_relation', (col('sum(QtdEscolasAtendidas)') / col('sum(pop_21)')) * 100)

# Ordenando os dados pelo percentual de relação, do maior para o menor
state_data = state_data.orderBy(col('percent_relation').desc())

# Convertendo para um DataFrame pandas para melhor manipulação com matplotlib
state_data_pd = state_data.toPandas()

# Função para formatar os números do eixo y em porcentagem
def to_percent(y, position):
    s = str(100 * y)
    return s + '%'

formatter = FuncFormatter(to_percent)

# Definindo o gráfico
fig, ax = plt.subplots(figsize=(14, 7))

# Barras para relação percentual
ax.bar(state_data_pd['uf_code'], state_data_pd['percent_relation'], color='green', edgecolor='white', label='Relação Percentual')

# Configuração do gráfico
ax.set_xlabel('Estados (UF)', fontweight='bold', fontsize=15)
ax.set_ylabel('Relação Percentual', fontweight='bold')
ax.set_title('Relação Percentual entre Escolas Atendidas e População por Estado', fontweight='bold')
ax.yaxis.set_major_formatter(formatter)
ax.set_xticklabels(state_data_pd['uf_code'], rotation=45)

fig.tight_layout()
plt.show()

```
